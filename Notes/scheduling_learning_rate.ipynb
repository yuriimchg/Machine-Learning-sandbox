# Adapting learning rate

Learning rate $\alpha$ responses for the gradient step. Too large $\alpha$ causes growing losses or missing the local/global minima, too tiny $\alpha$ requires too many steps for convergence. Andrew Ng in the Coursera Machine Learning course recommends to plot the loss function after every training step and, depending on the curve, make $\alpha$ three times larger / smaller.

$$
    W_{ij} = \alpha \times \frac{dJ}{dx}
$$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(-100,100)
y_1 = np.square(x - 40) + 40
y_3 = 1 / (x + 101)
y_2 = 1000 / (x + 101) + np.sin(x) + 101
y_4 = 50 /(x + 181) 
```

```python
plt.figure(figsize=(11,7))
plt.subplots_adjust(hspace=0.4)

plt.subplot(221)
plt.title(("Too large learning rate"))
plt.xlabel("Num_epochs")
plt.ylabel("Losses")
plt.plot(0,-10000)
plt.plot(x, y_1, c='r')
plt.gca().axes.xaxis.set_ticklabels([])
plt.gca().axes.yaxis.set_ticklabels([])

plt.subplot(222)
plt.title(("Still too large learning rate"))
plt.xlabel("Num_epochs")
plt.ylabel("Losses")
plt.plot(x, y_2, c='r')
plt.plot(0, -100)
plt.gca().axes.xaxis.set_ticklabels([])
plt.gca().axes.yaxis.set_ticklabels([])

plt.subplot(223)
plt.title(("Good enough learning rate"))
plt.xlabel("Num_epochs")
plt.ylabel("Losses")
plt.plot(x, y_3, c='g')
plt.gca().axes.xaxis.set_ticklabels([])
plt.gca().axes.yaxis.set_ticklabels([])

plt.subplot(224)
plt.title(("Too small learning rate"))
plt.xlabel("Num_epochs")
plt.ylabel("Losses")
plt.plot(x, y_4, c='r')
plt.plot(0,0)
plt.gca().axes.xaxis.set_ticklabels([])
plt.gca().axes.yaxis.set_ticklabels([])

plt.show()
```

<p style='font-size:18px; font-family:Impact;'>
 
 There is also a good idea to make learning rate alpha decrease. There are some possible mechanisms:
 <ol>
    <li><p>On every step.</p>
        <p>Let learning rate be $\alpha$, decay $\gamma$ is learning rate divided by number $n$ of epochs and use Nesterov accelerate gradient. Then for $i$-th step we can set  
            $$
                \alpha_{i+1} = \frac{\alpha_{i}}{1+\gamma_{i}},{~~~} where {~~} \gamma_{i}=\frac{\alpha_{i}}{n}
            $$
      </p>
    </li>
    <li>After pre-defined step. For example on every 100th epoch decrease $\alpha$ by 10%</li>
    <li>Change $\alpha$ manually (pause training, change the LR and continue)</li>
 </ol>
    
</p>

All of these techniques can be used with Keras LearningRateScheduler
