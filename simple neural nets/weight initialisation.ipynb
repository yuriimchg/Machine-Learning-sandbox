```python
import numpy as np
import matplotlib.pyplot as plt
```

 <center>
     <h2>Normal distribution: $ p(x) = \frac{1}{\sqrt{2\pi\sigma^2} }e^{-\frac{(x-\mu)^2}{2{\sigma^2}}}$ </h2>
        <br>

   <h2> Mean: $ \mu = 0 $</h2>
   <h2> Standart deviation: $ \sigma $ </h2>
</center>


```python
W_in = 64
W_out = 32
limit = np.sqrt(1 / float(W_in))
W = np.random.normal(0.0, limit, size=(W_in, W_out))
```

```python
plt.figure(figsize=(20,10))
plt.hist(W)
plt.show()
```

 <center><h2> Uniform distribution </h2></center>

```python
W = np.random.uniform(low=-0.05, high=0.05, size=(W_in, W_out))
```

```python
plt.figure(figsize=(20,10))
plt.hist(W)
plt.show()
```

```python
limit = np.sqrt(6 / float(W_in + W_out))
W = np.random.uniform(low=-limit, high=limit, size=(W_in, W_out))
```

 <center><h2> Uniform distribution with stronger restrictions on limit</h2></center>

```python
plt.figure(figsize=(20,10))
plt.hist(W)
plt.show()
```
